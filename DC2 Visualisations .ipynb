```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the preprocessed CSV file into a pandas DataFrame
df = pd.read_csv('burglaries_preprocessed.csv')

# Convert the 'Month' column to datetime type
df['Month'] = pd.to_datetime(df['Month'])

# Extract the year from the 'Month' column
df['Year'] = df['Month'].dt.year

# Group the data by year and ward and count the number of crimes in each combination
crimes_by_year_ward = df.groupby(['Year', 'Ward']).size().unstack()

# Plotting the bar chart
crimes_by_year_ward.plot(kind='bar', figsize=(12, 6))
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.title('Number of Crimes in Each Ward by Year')
plt.legend(title='Ward')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the preprocessed CSV file into a pandas DataFrame
df = pd.read_csv('burglaries_preprocessed.csv')

# Convert the 'Month' column to datetime type
df['Month'] = pd.to_datetime(df['Month'])

# Extract the year from the 'Month' column
df['Year'] = df['Month'].dt.year

# Group the data by year and count the number of crimes in each year
crimes_by_year = df['Year'].value_counts().sort_index()

# Plotting the bar chart
crimes_by_year.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.title('Number of Crimes by Year')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the preprocessed CSV file into a pandas DataFrame
df = pd.read_csv('burglaries_preprocessed.csv')

# Group the data by Ward and LSOA, and count the number of crimes
ward_lsoa_counts = df.groupby(['Ward', 'LSOA name']).size().reset_index(name='Crime Count')

# Sort the data by Crime Count in descending order
ward_lsoa_counts = ward_lsoa_counts.sort_values(by='Crime Count', ascending=False)

# Select the top 5 wards with the highest number of crimes
top_wards = ward_lsoa_counts['Ward'].unique()[:5]

# Filter the data for the top 5 wards
top_ward_lsoa_counts = ward_lsoa_counts[ward_lsoa_counts['Ward'].isin(top_wards)]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_ward_lsoa_counts['LSOA name'], top_ward_lsoa_counts['Crime Count'])
plt.xlabel('LSOA')
plt.ylabel('Number of Crimes')
plt.title('Number of Crimes in LSOA for Top 5 Wards')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the preprocessed CSV file into a pandas DataFrame
df = pd.read_csv('burglaries_preprocessed.csv')

# Group the data by Ward and LSOA, and count the number of crimes
ward_lsoa_counts = df.groupby(['Ward', 'LSOA name']).size().reset_index(name='Crime Count')

# Sort the data by Crime Count in descending order
ward_lsoa_counts = ward_lsoa_counts.sort_values(by='Crime Count', ascending=False)

# Select the top 5 wards with the highest number of crimes
top_wards = ward_lsoa_counts['Ward'].unique()[:5]

# Filter the data for the top 5 wards
top_ward_lsoa_counts = ward_lsoa_counts[ward_lsoa_counts['Ward'].isin(top_wards)]

# Plotting the bar chart
plt.figure(figsize=(10, 6))

# Iterate over the top wards
for ward in top_wards:
    # Filter the data for the current ward
    ward_data = top_ward_lsoa_counts[top_ward_lsoa_counts['Ward'] == ward]
    
    # Get the LSOA names and crime counts
    lsoa_names = ward_data['LSOA name']
    crime_counts = ward_data['Crime Count']
    
    # Plot the bar chart for the current ward
    plt.bar(lsoa_names, crime_counts, label=ward)

plt.xlabel('LSOA')
plt.ylabel('Number of Crimes')
plt.title('Number of Crimes in LSOA for Top 5 Wards')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the preprocessed CSV file into a pandas DataFrame
df = pd.read_csv('burglaries_preprocessed.csv')

# Count the occurrences of each outcome category
outcome_counts = df['Last outcome category'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))

# Plot the pie chart
plt.pie(outcome_counts.values, labels=None, autopct='%1.1f%%', startangle=90)

plt.title('Distribution of Outcomes')

# Add legend with labels at the bottom
plt.legend(labels=outcome_counts.index, loc='lower center', bbox_to_anchor=(0.5, -0.1), ncol=3)

plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("burglaries_preprocessed.csv")

# Group the data by outcome category and count the number of crimes in each category
outcome_counts = data.groupby("Last outcome category")["Crime ID"].count()

# Keep the 6 highest values and group the rest in an "Other" category
other_count = outcome_counts.nsmallest(len(outcome_counts)-6).sum()
outcome_counts = outcome_counts.nlargest(3).append(pd.Series(other_count, index=["Other"]))

# Plot the pie chart
plt.pie(outcome_counts, labels=outcome_counts.index, autopct='%1.1f%%')

# Add a title
plt.title("Crime Outcomes")

# Add a legend at the bottom
plt.legend(labels=outcome_counts.index, loc='lower center', bbox_to_anchor=(0.5, -0.1), ncol=3)

# Show the plot
plt.show()
```
