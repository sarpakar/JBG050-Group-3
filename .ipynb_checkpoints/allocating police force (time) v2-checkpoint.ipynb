```python
# is a function that gives a ratio column in the df
def breuk(df):
    lst = []
    for index, row in df.iterrows():
        if df['Timeslot'][index] == '08.00 - 09.00' or df['Timeslot'][index] == '21.00 - 22.00':
            value = df['Amount_of_sas'][index]
            lst.append(value)
        else:
            value = df['Amount_of_sas'][index] * 2
            lst.append(value)

    df['Amount_of_sas_time'] = lst
    lst_breuk = []
    total = sum(lst)
    for index, row in df.iterrows():
        value = df['Amount_of_sas_time'][index]
        breuk = value / total
        lst_breuk.append(breuk)
    df['Ratio_sas'] = lst_breuk
    return df

```
