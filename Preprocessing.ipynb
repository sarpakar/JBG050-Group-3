```python
import pandas as pd
import csv
import glob
import random
```

```python
df = pd.concat([pd.read_csv(f) for f in glob.glob('data_2010_to_2023/'+'*.csv')])
df
```

```python
df_burglary = df[df['Crime type'] == 'Burglary']
df_burglary = df_burglary.reset_index(drop = True)
df_burglary
```

```python
df_burglary.dropna(subset = ['LSOA name','LSOA code'], inplace=True)

```

```python

df_barnet = df_burglary.loc[df_burglary['LSOA name'].str.contains("Barnet", case=False)]
df_barnet = df_barnet.reset_index(drop = True)

```

```python
# convert the LSOA encoding to the 2021 version

lsoa_raw = 'LSOA_(2011)_to_LSOA_(2021).csv'

lsoa = pd.read_csv(lsoa_raw)

# filter the data to include only the LSOA codes in Barnet
barnet_lsoas = lsoa[lsoa['LAD22NM'] == 'Barnet']


df_lsoa_split = barnet_lsoas[barnet_lsoas['LSOA11CD'].duplicated(keep=False)]#value_counts()

df_lsoa_split.sort_values(by=['LSOA11CD'])
# lsoa's that changed from 11 to 21
```

```python
lsoa_splitted_lst = list(df_lsoa_split['LSOA11CD'].unique())

lsoa_splitted_dct = {'E01000125':['E01033913', 'E01033914'],
                   'E01000148': ['E01033923', 'E01033924'],
                   'E01000155': ['E01033916', 'E01033921', 'E01033922'],
                   'E01000262': ['E01033911', 'E01033912'],
                   'E01033572': ['E01033915', 'E01033919', 'E01033920', 'E01033925'],
                   'E01033573': ['E01033917','E01033918']}

df_replace = df_barnet[df_barnet['LSOA code'].isin(lsoa_splitted_lst)]
# df with rows to replace inside df_barnet
```

```python
old_lsoa = list(df_barnet['LSOA code'].values)
new_lsoa = []

for idx, row in df_barnet.iterrows():
    if row['LSOA code'] in lsoa_splitted_lst:
        new_lsoa.append(random.choice(lsoa_splitted_dct[row['LSOA code']]))
        # to reassign crimes that happened with the 2011 encoding to the 2021 one,
        # we choose a random code from the ones that the old code was split into and assign the crime there
        # so we have all crimes assigned to a code with the current naming scheme
    else: 
        new_lsoa.append(df_barnet.iloc[idx]['LSOA code'])   
        
# new_lsoa list contains the unchanged lsoa codes and the changed ones
# old_lsoa list contains the unchanged codes 
# in the end we replace old with new codes 
```

```python
df_barnet['LSOA code'].replace(old_lsoa, new_lsoa, inplace=True)
```

```python
# check if all LSOA codes fall within our 2021 LSOA encoding
df_barnet[df_barnet["LSOA code"].isin(list(barnet_lsoas['LSOA21CD'].unique())) == False]

```

```python
# WARDS

# opens a dataframe where you have all the information of the LSOA's and to which wards they belong. 

df_t = pd.read_csv('Area_(2021)_to_Ward_(2022).csv')
df_LSOA21NM = df_t.loc[df_t['LSOA21NM'].str.contains("Barnet", case=False)]
df_LSOA21NM

# did some test and the wards align with the readings

df_LSOA21NM.reset_index(drop = True, inplace=True)
```

```python
df_LSOA21NM
```

```python
df_LSOA21NM["WD22NM"].unique()
```

```python
# makes a dictionary where every LSOA is a key and the ward of the LSOA is the value

dct_LSOA_to_WARD = {}
for index, row in df_LSOA21NM.iterrows():
    key = row['LSOA21CD']
    value = df_LSOA21NM["WD22NM"][index]
    dct_LSOA_to_WARD[key] = value
    
dct_LSOA_to_WARD
    
    
```

```python
# appends a new column called ward to the df_original.
# it first looks at the LSOA code an looks that code up un the dictionary which gives the wards of the LSOA
# and it puts that ward in the ward column

wards_lst = []
for index, row in df_barnet.iterrows():
    LSOA = row['LSOA code']
    ward = dct_LSOA_to_WARD[LSOA] 
    wards_lst.append(ward)
    
        
df_barnet['Ward'] = wards_lst
```
