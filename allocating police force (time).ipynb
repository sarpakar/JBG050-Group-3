```python
import pandas as pd
import csv
import glob
import random
import re
import seaborn as sns
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# opens csv file
df_sas = pd.read_csv('stop_and_search_preprocessed.csv')
df_sas
```

```python
# makes a column timeslot of the time slot and it makes a column timeslot index.
timeslot_lst = []
timeslot_index = []
for index, row in df_sas.iterrows():
    date_time = df_sas['Date'][index].split('T')
    time = date_time[1].split(':')
    hour = time[0]
    if hour == '00':
        timeslot_lst.append(('00.00 - 01.00'))
        timeslot_index.append(0)
    if hour == '01':
        timeslot_lst.append(('01.00 - 02.00'))
        timeslot_index.append(1)
    if hour == '02':
        timeslot_lst.append(('02.00 - 03.00'))
        timeslot_index.append(2)
    if hour == '03':
        timeslot_lst.append(('03.00 - 04.00'))
        timeslot_index.append(3)
    if hour == '04':
        timeslot_lst.append(('04.00 - 05.00'))
        timeslot_index.append(4)
    if hour == '05':
        timeslot_lst.append(('05.00 - 06.00'))
        timeslot_index.append(5)
    if hour == '06':
        timeslot_lst.append(('06.00 - 07.00'))
        timeslot_index.append(6)
    if hour == '07':
        timeslot_lst.append(('07.00 - 08.00'))
        timeslot_index.append(7)
    if hour == '08':
        timeslot_lst.append(('08.00 - 09.00'))
        timeslot_index.append(8)
    if hour == '09':
        timeslot_lst.append(('09.00 - 10.00'))
        timeslot_index.append(9)
    if hour == '10':
        timeslot_lst.append(('10.00 - 11.00'))
        timeslot_index.append(10)
    if hour == '11':
        timeslot_lst.append(('11.00 - 12.00'))
        timeslot_index.append(11)
    if hour == '12':
        timeslot_lst.append(('12.00 - 13.00'))
        timeslot_index.append(12)
    if hour == '13':
        timeslot_lst.append(('13.00 - 14.00'))
        timeslot_index.append(13)
    if hour == '14':
        timeslot_lst.append(('14.00 - 15.00'))
        timeslot_index.append(14)
    if hour == '15':
        timeslot_lst.append(('15.00 - 16.00'))
        timeslot_index.append(15)
    if hour == '16':
        timeslot_lst.append(('16.00 - 17.00'))
        timeslot_index.append(16)
    if hour == '17':
        timeslot_lst.append(('17.00 - 18.00'))
        timeslot_index.append(17)
    if hour == '18':
        timeslot_lst.append(('18.00 - 19.00'))
        timeslot_index.append(18)
    if hour == '19':
        timeslot_lst.append(('19.00 - 20.00'))
        timeslot_index.append(19)
    if hour == '20':
        timeslot_lst.append(('20.00 - 21.00'))
        timeslot_index.append(20)
    if hour == '21':
        timeslot_lst.append(('21.00 - 22.00'))
        timeslot_index.append(21)
    if hour == '22':
        timeslot_lst.append(('22.00 - 23.00'))
        timeslot_index.append(22)
    if hour == '23':
        timeslot_lst.append(('23.00 - 00.00'))
        timeslot_index.append(23)
        
df_sas['Timeslot'] = timeslot_lst
df_sas['Timeslot index'] = timeslot_index
df_sas
```

```python
# groups the SAS by timeslot index and ward
group_allocation_sas = df_sas.groupby(['Timeslot index', 'NAME']).size().reset_index(name='Amount_of_sas')
group_allocation_sas

# groups the SAS by timeslot index and ward but only where the outcome of the SAS ended in a arrest or suspect arrest.
df_allocation = df_sas.copy()
df_allocation = df_allocation[df_allocation['Outcome'].isin(['Arrest', 'Suspect arrest'])]  #only 2331 rows
group_allocation_sas_arrest = df_allocation.groupby(['Timeslot index', 'NAME']).size().reset_index(name='Amount_of_sas')

group_allocation_sas_arrest
```

```python
#makes a new column group timeslots which groups timeslots of two hours together.
timeslot_lst = []
for index, row in df_sas.iterrows():
    Timeslot_index = df_sas['Timeslot index'][index]
    if Timeslot_index == 0 or Timeslot_index == 1:
        timeslot_lst.append(('00.00 - 02.00'))
    if Timeslot_index == 2 or Timeslot_index == 3:
        timeslot_lst.append(('02.00 - 04.00'))
    if Timeslot_index == 4 or Timeslot_index == 5:
        timeslot_lst.append(('04.00 - 06.00'))
    if Timeslot_index == 6 or Timeslot_index == 7:
        timeslot_lst.append(('06.00 - 08.00'))
    if Timeslot_index == 8 or Timeslot_index == 9:
        timeslot_lst.append(('08.00 - 10.00'))
    if Timeslot_index == 10 or Timeslot_index == 11:
        timeslot_lst.append(('10.00 - 12.00'))
    if Timeslot_index == 12 or Timeslot_index == 13:
        timeslot_lst.append(('12.00 - 14.00'))
    if Timeslot_index == 14 or Timeslot_index == 15:
        timeslot_lst.append(('14.00 - 16.00'))
    if Timeslot_index == 16 or Timeslot_index == 17:
        timeslot_lst.append(('16.00 - 18.00'))
    if Timeslot_index == 18 or Timeslot_index == 19:
        timeslot_lst.append(('18.00 - 20.00'))
    if Timeslot_index == 20 or Timeslot_index == 21:
        timeslot_lst.append(('20.00 - 22.00'))
    if Timeslot_index == 22 or Timeslot_index == 23:
        timeslot_lst.append(('22.00 - 00.00'))


df_sas['Grouped timeslots'] = timeslot_lst
df_sas
```

```python
# grouped timeslots togeter based on the 2 hour grouped timeslot and the ward
group_timeslots_sas = df_sas.groupby(['Grouped timeslots', 'NAME']).size().reset_index(name='Amount_of_sas')
group_timeslots_sas
```
